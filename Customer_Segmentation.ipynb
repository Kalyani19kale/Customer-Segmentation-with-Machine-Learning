```python
numerical_columns = df.select_dtypes(include=['number']).columns
print(numerical_columns)
```

```python
numerical_analysis=df[numerical_columns].describe()
print(numerical_analysis)
```

```python
for column in numerical_columns:
  print(f"Column Name: {column}")
  mean_val=df[column].mean()
  median_val=df[column].median()
  mode_val=df[column].mode()
  std_dev =df[column].std()
  variance =df[column].var()
  print(f"Mean: {mean_val}")
  print(f"Median: {median_val}")
  print(f"Mode: {mode_val}")
  print(f"Standard Deviation: {std_dev}")
  print(f"Variance: {variance}")
```

```python
#RFM
data_columns=['Quantity', 'InvoiceDate','Price', 'Customer ID']
for column in data_columns:
  if column in df.columns:
     print(f"columns : {column}")
     #gives unique value in that column
     unique_levels=df[column].unique()
     print(f"Unique levels: {unique_levels}")
     #give for that unique value how much number of value exist eg high=80 low=30
     level_distribution=df[column].value_counts()
     print(f"level distribution :\n{level_distribution}")
     #gives max value for that column
     most_frequent_level = level_distribution.idxmax()
     print(f"Most Frequent Level: {most_frequent_level}")

```
